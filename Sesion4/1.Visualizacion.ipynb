```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Grafico de linea 
x = [1,2,3,4,5]
y = [2,3,4,5,9]

plt.plot(x,y)
plt.xlabel("eje x")
plt.ylabel("eye y")
plt.title("Grafico de linea")

plt.show()
```

```python
#Grafico de dispersion 
x = [1,2,3,4,5]
y = [2,3,4,5,9]

plt.scatter(x,y)
plt.xlabel("eje x")
plt.ylabel("eye y")
plt.title("Grafico de dispersion")

plt.show()
```

```python
#Histograma 
import numpy as np 
datos = np.random.rand(1000)

plt.hist(datos, bins=30, color='skyblue', edgecolor='red')
plt.xlabel("eje x")
plt.ylabel("eye y")
plt.title("Histograma")
plt.show()
```

```python
#Grafico de Barras 

categoria = ['A','B','C']
valor = [150, 230, 180]

plt.bar(categoria, valor, color='purple')
plt.xlabel("eje x")
plt.ylabel("eye y")
plt.title("Grafico de Barras")
plt.show()             
```
